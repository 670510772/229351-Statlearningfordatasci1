Report the test accuracy of your best model.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier, plot_importance as plot_xgb
from lightgbm import LGBMClassifier, plot_importance as plot_lgbm

X_train = pd.read_csv('X_train.csv', header=None)
y_train = pd.read_csv('y_train.csv', header=None).values.ravel()
X_test = pd.read_csv('X_test.csv', header=None)
y_test = pd.read_csv('y_test.csv', header=None).values.ravel()
models_config = [
    {
        'name': 'AdaBoost',
        'estimator': AdaBoostClassifier(algorithm='SAMME'),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 1.0]
        }
    },
    {
        'name': 'XGBoost',
        'estimator': XGBClassifier(eval_metric='logloss'),
        'params': {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.05, 0.1]
        }
    },
    {
        'name': 'LightGBM',
        'estimator': LGBMClassifier(verbose=-1),
        'params': {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.05, 0.1]
        }
    }
]

best_final_model = None
target_accuracy = 0.83
```

```python
results_list = []

for cfg in models_config:
    print(f"{cfg['name']}...")

    gscv = GridSearchCV(
        estimator=cfg['estimator'],
        param_grid=cfg['params'],
        cv=5,
        scoring='accuracy'
    )
    gscv.fit(X_train, y_train)

    oof_accuracy = gscv.best_score_

    y_pred = gscv.best_estimator_.predict(X_test)
    test_acc = accuracy_score(y_test, y_pred)

    results_list.append({
        'Model': cfg['name'],
        'OOF Accuracy': oof_accuracy,
        'Test Accuracy': test_acc,
        'Best Params': gscv.best_params_
    })

    print(f"-> {cfg['name']}  (OOF: {oof_accuracy:.4f}, Test: {test_acc:.4f})")

df_comparison = pd.DataFrame(results_list).sort_values(by='Test Accuracy', ascending=False)

print("\n" + "="*50)
print("="*50)
print(df_comparison)
```

For each model, plot the feature importances

```python
import numpy as np
import matplotlib.pyplot as plt

best_ad = gscv.best_estimator_
importances = best_ad.feature_importances_
feature_names = X_train.columns
indices = np.argsort(importances)

plt.figure(figsize=(10, 8))
plt.title('Feature Importances: AdaBoost Classifier')
plt.barh(range(len(indices)), importances[indices], color='skyblue', align='center')
plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
plt.xlabel('Relative Importance Score', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
